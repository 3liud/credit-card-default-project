```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# rednder plotting automatically
%matplotlib inline

#additional ploting functionality
import matplotlib as mpl
# high resolution figures
mpl.rcParams['figure.dpi'] = 500
```

```python
# load data
df_clean_2 = pd.read_csv('../Data/df_clean_2.csv')
```

```python
# histogram on limit balance and age
df_clean_2[['LIMIT_BAL', 'AGE']].hist()
```

```python
# tabular report of summary statistics 
df_clean_2[['LIMIT_BAL', 'AGE']].describe()
```

- minimum creadit limit is 10,000
- age is reasonably distributed and the minimum age is 21 years.
- credit limit is reasonable as it should be above 0 while age is reasonable as no under ages shoudl get loans

```python
# value counts on education 
df_clean_2['EDUCATION'].value_counts()
```

- 1 = Graduate school
- 2 = University
- 3 = high school
- 4 = others

- 0, 5, 6 are undocumneted from the data dictionary (we wll lump them with the potehrs category)

```python
# lumping undocumnted education levels with others category
df_clean_2['EDUCATION'].replace(to_replace=[0, 5, 6], value=4, inplace=True)
df_clean_2['EDUCATION'].value_counts()
```

```python
# value counts for marriage
df_clean_2['MARRIAGE'].value_counts()
```

- 1 = Married
- 2 = single
- 3 = others
- 0 = undocumneted from the data dictionary
- we lump the undocumented in the other category

```python
# Lump the undocumented category in otehr category
df_clean_2['MARRIAGE'].replace(to_replace= 0, value=3, inplace=True)
df_clean_2['MARRIAGE'].value_counts()
```
